# 清理数据

基于以上数据评估，所需的数据清理步骤如下：  
1.InvoiceDate数据类型转换为日期时间  
2.CustomerID数据类型转换应为字符串  
3.剔除Description存在缺失值行（CustomerID不影响后续数据分析，暂不进行处理）  
4.Country变量值“USA”替换为“United States”  
5.Country变量值“UK”“U.K.”替换为“United Kingdom”  
6.剔除Quantity<0观察值行  
7.剔除UnitPrice<0观察值行  

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

InvoiceDate数据类型转换为日期时间

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data["InvoiceDate"]
```

CustomerID数据类型转换应为字符串

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)
cleaned_data["CustomerID"]
```

删除CustomerID结尾的0

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].str.slice(0, -2)
cleaned_data["CustomerID"]
```

剔除Description存在缺失值行

```python
cleaned_data.dropna(subset = ["Description"], inplace = True)
```

操作验证

```python
cleaned_data["Description"].isnull().sum()
```

Country变量值“USA”替换为“United States”

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"USA": "United States"})
```

操作验证

```python
len(cleaned_data[cleaned_data["Country"] == "USA"])
```

Country变量值“UK”“U.K.”替换为“United Kingdom”

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"UK": "United States", "U.K.": "United States"})
```

```python
print(len(cleaned_data[cleaned_data["Country"] == "UK"]))
print(len(cleaned_data[cleaned_data["Country"] == "UK"]))
```

剔除Quantity<0观察值行

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] >= 0]
```

```python
len(cleaned_data[cleaned_data["Quantity"] < 0])
```

剔除UnitPrice<0观察值行

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >= 0]
```

```python
len(cleaned_data[cleaned_data["UnitPrice"] < 0])
```

# 保存清理数据

```python
cleaned_data.to_csv(r"F:\Python exercise\Jupyter\e_commerce_cleaned.csv", index = False)
#不导出索引
```
